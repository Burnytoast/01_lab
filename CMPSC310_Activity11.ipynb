## Binomial Distribution

Binomial disrtibution is defined as "Binomial distribution is calculated by multiplying the probability of success raised to the power of the number of successes and the probability of failure raised to the power of the difference between the number of successes and the number of trials." The output seems to be floating right around 50 +/- 10, over the course of 100 instances.

```python
import numpy as np

n=100 # number of trials
p=0.5 # probability of success
s=100 # size
binomial = np.random.binomial(n,p,s)
print(binomial)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 100])
plt.plot(binomial,linewidth=2, color='r')
plt.show()
```

## Uniform Distribution

Uniform distributions are described as "Uniform distributions are probability distributions with equally likely outcomes. In a discrete uniform distribution, outcomes are discrete and have the same probability. In a continuous uniform distribution, outcomes are continuous and infinite. In a normal distribution, data around the mean occur more frequently." Within the 100 instances seen, the values seemed to bounce from 9.5 to 1.

```python
import numpy as np

uniform = np.random.uniform(low=1, high=10,size=100)
print(uniform)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 10])
plt.plot(uniform,linewidth=2, color='r')
plt.show()
```

## Normal Distribution

Normal distribution is described as "What Is a Normal Distribution? Normal distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. In graphical form, the normal distribution appears as a "bell curve"." 

```python
import numpy as np
mu, sigma = 0, 0.1 # mean and standard deviation
normal = np.random.normal(mu, sigma, 100)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 1])
plt.plot(normal,linewidth=2, color='r')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian is normal distribution
X, y = load_iris(return_X_y=True) # https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```
